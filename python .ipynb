```
correct_username = "admin"
correct_password = "1234"

attempts = 0

while attempts < 3:
    username = input("Enter username: ")
    password = input("Enter password: ")

    if username == correct_username and password == correct_password:
        print("Login successful! Welcome,", username)
        break

    elif username == correct_username and password != correct_password:
        print("Incorrect password. Try again.")

    else:
        print("User not found.")

    attempts += 1

if attempts == 3:
    print("Account locked.")
```

A

```
pinNumber = 2221     # Correct PIN
balance = 6000      # Account balance

userPin = int(input("Enter PIN: "))

if userPin != pinNumber:
    print("Wrong PIN!")
else:
    request = float(input("Enter withdrawal amount: "))

    if request <= 500:
        print("Amount must be greater than 500")
    elif request > balance:
        print("Insufficient balance")
    else:
        balance = balance - request
        print("Successful withdrawal")
        print("Remaining balance:", balance)
```

B

```
pinNumber = 6678
balance = 9999
attempts = 5
status = True

while attempts > 0:
    userPin = int(input("Enter PIN: "))

    if userPin == pinNumber:
        print("PIN correct!")

        request = float(input("Enter withdrawal amount: "))

        if request <= 500:
            print("Amount must be greater than 500")

        elif request > balance:
            print("Insufficient balance")

        else:
            balance = balance - request
            print("Withdrawal successful")
            print("Remaining balance:", balance)

        break
    else:
        attempts -= 1

        if attempts == 0:
            print("Card blocked!")
            status = False
        else:
          print("Wrong pin. Attempts remaining:", attempts)
```

```
correct_username = "MSH"
correct_password = "0775"
is_active = True
attempts = 5

while attempts > 0:

    username = input("Enter username: ")
    password = input("Enter password: ")

    if is_active == False:
        print("Account is disabled.")
        break

    elif username != correct_username or password != correct_password:
        attempts -= 1
        print("Login failed. Attempts left:", attempts)

    else:
        print("Login successful!")
        break

if attempts == 0:
    print("Account disabled.")
```
